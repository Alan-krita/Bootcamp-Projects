```python
students = []

for i in range(5):
    name = input("Enter student name: ")
    study_hours = int(input("Enter study hours: "))
    attendance = int(input("Enter attendance: "))
    marks = int(input("Enter marks: "))

    student = {
        "name": name,
        "study_hours": study_hours,
        "attendance": attendance,
        "marks": marks
    }

    students.append(student)


for s in students:
    print("\nName:", s["name"])

    if s["marks"] > 75:
        print("Performance: Strong")
    elif s["marks"] >= 40:
        print("Performance: Need Improvement")
    else:
        print("Performance: Fail")

    if s["attendance"] < 75:
        print("Warning: Low Attendance")

    if s["study_hours"] < 4:
        print("Warning: Need More Study Time")
```
