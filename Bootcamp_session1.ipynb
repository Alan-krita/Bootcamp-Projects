```python
# Visualise dataset using matplotlib library

import pandas as pd

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [12000, 15000, 18000, 17000, 22000, 25000,
              28000, 27000, 26000, 30000, 35000, 40000],
    'Profit': [2500, 3000, 3500, 3200, 4500, 5000,
               6000, 5800, 5500, 6500, 8000, 9500],
    'Customers': [150, 180, 220, 210, 260, 290,
                  330, 320, 310, 350, 420, 500],
    'Returns': [12, 15, 18, 14, 20, 25,
                22, 21, 19, 24, 30, 35]
}

df = pd.DataFrame(data)

print(df.head())
```

```python
# Line Chart
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Sales'], marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(8,5))

plt.bar(df['Month'], df['Profit'])

plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit ($)")
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8,5))

plt.scatter(df['Customers'], df['Sales'])

plt.title("Customers vs Sales")
plt.xlabel("Number of Customers")
plt.ylabel("Sales")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8,5))

plt.hist(df['Sales'], bins=6)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pie chart
quarters = {
    'Q1': sum(df['Sales'][0:3]),
    'Q2': sum(df['Sales'][3:6]),
    'Q3': sum(df['Sales'][6:9]),
    'Q4': sum(df['Sales'][9:12])
}

plt.figure(figsize=(7,7))

plt.pie(quarters.values(),
        labels=quarters.keys(),
        autopct='%1.1f%%')

plt.title("Sales Contribution by Quarter")
plt.show()
```

```python
# Horizontal Bar Chart
top = df.sort_values('Sales', ascending=False)

plt.figure(figsize=(8,5))

plt.barh(top['Month'], top['Sales'])

plt.title("Sales by Month")
plt.xlabel("Sales")
plt.show()
```

```python
# Box plot
plt.figure(figsize=(8,5))

plt.boxplot([df['Sales'],
             df['Profit'],
             df['Returns']],
            labels=['Sales','Profit','Returns'])

plt.title("Distribution of Business Metrics")
plt.show()
```
