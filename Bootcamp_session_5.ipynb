```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df = pd.read_csv(url)

# 1. Range of years
min_year = df['year'].min()
max_year = df['year'].max()

print("Range of Years:", min_year, "-", max_year)

# 2. Total number of countries
total_countries = df['country'].nunique()
print("Total Number of Countries:", total_countries)

# 3. Number of continents
total_continents = df['continent'].nunique()
print("Number of Continents:", total_continents)

#List all continents
print("\nContinents:")
print(df['continent'].unique())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)

# Check column names
print(df.columns)

# Calculate average life expectancy by continent
avg_life_exp = df.groupby('continent')['lifeExp'].mean()

# Create bar chart
plt.figure(figsize=(8, 5))
avg_life_exp.plot(kind='bar')

plt.title('Average Life Expectancy by Continent')
plt.xlabel('Continent')
plt.ylabel('Average Life Expectancy')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

# Filter for the three countries
countries = ['India', 'China', 'United States']
filtered_df = df[df['country'].isin(countries)]

# Create the line chart
plt.figure(figsize=(8, 4))

for country in countries:
    country_data = filtered_df[filtered_df['country'] == country]
    plt.plot(
        country_data['year'],
        country_data['lifeExp'],
        marker='*',
        label=country
    )

plt.title('Life Expectancy Over Time')
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Years)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

# Create scatter plot
plt.figure(figsize=(9, 5))
plt.scatter(df['gdpPercap'], df['lifeExp'], alpha=0.6)

plt.title('GDP per Capita vs Life Expectancy')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy (Years)')
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

# Filter data for 2007
df_2007 = df[df['year'] == 2007]

# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df_2007['gdpPercap'], df_2007['lifeExp'], alpha=0.7)

plt.xscale('log')  # Makes GDP distribution easier to visualize

plt.title('GDP per Capita vs Life Expectancy (2007)')
plt.xlabel('GDP per Capita (Log Scale)')
plt.ylabel('Life Expectancy (Years)')
plt.grid(True)

plt.show()
```
