```python
# Student performance analyzer

import pandas as pd
import numpy as np

# Read session4_dummy_students.csv
df=pd.read_csv("session4_dummy_students_data(bootcamp).csv.xls")

# Create a result column: Pass if average>=40, othwerwise fail
def get_result(avg):
  if avg>=40:
    return "Pass"
  else:
     return "Fail"

# Create an average column using maths,science and english
df['average']= df[["math","science","english"]].mean(axis=1)
df['result']=df['average'].apply(get_result)

print("Full data:")
print(df)

# Print class average,highest average and lowest average
print("Class average:",df['average'].mean())
print("Highest average:",df['average'].max())
print("Lowest average:",df['average'].min())

# Print top students
Top_students=df[df['average']>=85]
print("Top students:",Top_students[["name","average"]])

# Print students with attendance below 75
low_attendance = df[df["attendance_percent"] < 75]
print("Students with low attendance:",low_attendance[["name", "attendance_percent"]])

# Print students who submitted fewer than 6 assignments
low_assignments = df[df["assignments_submitted"] < 6]
print("Students with low assignments:",low_assignments[["name", "assignments_submitted"]])

```
